```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
def one_or_zero(x):
    return 1 if x > 3 else 0

def multiple_columns(x,y,z,k):
    a = x ** y
    b = y/z
    c = k ** k
    d = x+y+z+k
    number = ((1+a)+(2+b)+(3+c)+(4+d))
    denom = a+b+c+d
    return number/denom

def one_or_zero2(df):
    if df['sepal_length'] > 3:
        return df['sepal_width']*3
    return df['sepal_width']*2
```

```python
iris['new'] = iris['sepal_length'].apply(lambda x: one_or_zero2(x))
iris
```

```python
iris['test'] = iris['sepal_length'].map(lambda x :one_or_zero(x))
iris
```

```python
iris.apply(lambda x:multiple_columns(iris['sepal_length'],iris['sepal_width'],iris['petal_length'],iris['petal_width'])  )
```

```python
df2 = pd.DataFrame({ 'A': [1,2,3,4], 
                   'B': [10,20,30,40],
                   'C': [20,40,60,80]
                  }, 
                  index=['Row 1', 'Row 2', 'Row 3', 'Row 4'])

def custom_sum(row):
    return row.sum()
```

```python
df2['D'] = df2.apply(custom_sum, axis=1)
df2
```

```python
df2.loc['Row 5'] = df2.apply(custom_sum, axis=0)
df2
```

```python
def multiply_by_2(val):
    return val * 2

df2['E'] = df2['C'].apply(multiply_by_2)
df2
```

```python
df2['F'] = df2.apply(lambda x : x.sum(), axis=1)
df2
```

```python
df2['F2'] = df2['C'].apply(lambda x: x*2)
df2
```
