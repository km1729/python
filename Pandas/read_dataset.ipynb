```python
# from sklearn.datasets import load_iris
import sklearn.datasets as sdt
import pandas as pd
```

```python
# datasets을 import 해서 scikit에서 제공하는 데이터셋을 가져올 수 있다.
# data = sdt.load_breast_cancer()
# data = sdt.load_breast_cancer()
# data = sdt.load_diabetes()
# 데이터셋 https://scikit-learn.org/stable/datasets/toy_dataset.html

# skikit dataset은 bunch 이며, 딕셔너리와 같다
data = sdt.load_iris()
print(data.keys())
print(data['target_names'])
print(data['feature_names'])
```

```python
# 딕셔너리를 이용하여 판다 데이터프레임 만들기
# 데이터 키와 fature_name을 이용하여 데이터 프레임을 만든다
# 타겟 key를 이용하여 분류를 추가 한다.
# 타겟의 밸류와 타겟네임을 매칭 한다
df = pd.DataFrame(data=data.data,columns=data.feature_names)
df['target'] = pd.Series(data.target)
df['species'] = pd.Categorical.from_codes(data.target, data.target_names)
df
```
