```python
import math
class Point:
  def setdata(self, x, y):
    self.x = x
    self.y = y

def print_class(p):
  return print(f'point x, y = ({p.x}, {p.y})')

def distance_between_point(p):
  return math.sqrt(p.x**2 + p.y**2)
```

```python
blank = Point()
blank
```

```python
blank.x = 3.0
blank.y = 4.0
print(blank.x)
print(blank.y)
print(f"({blank.x} , {blank.y})")
```

```python
print_class(blank)
```

```python
print(distance_between_point(blank))
```
